```python
# Declaring dependencies
import pandas as pd
import numpy as np
```

```python
# CSV file path
csv_path = "../Resources/purchase_data.csv"
```

```python
# reading the csv file
purchase_data_df = pd.read_csv(csv_path, encoding = 'utf-8')
#purchase_data_df
```

```python
#checking if all columns are equal in lengh to avoid any errors while calculation

check = purchase_data_df.dropna(how='any')
#check.count()
```

```python
### Player Count

Player_count = len(purchase_data_df["SN"].unique().tolist())

count = pd.DataFrame({"Total Players":[Player_count]})
 
count
```

```python

#Unique Items as list
unique_items_df = len(purchase_data_df["Item Name"].unique().tolist())

#Total revenue
Total_revenue = purchase_data_df["Price"].sum()

Total_items = len(purchase_data_df)

#Average Purchase Price = Total revenue/Total items
Average_price = round(Total_revenue/Total_items, 2)


print("####Purchasing Analysis(Total)#####")


#print("Number of unique items are", len(unique_items_df))
#print("total revenue is", Total_revenue)
#print("Average purchase price is",Average_price)
#print()


summary_df = pd.DataFrame({"Number of Unique Items":[unique_items_df],
                           "Average Price":[Average_price], 
                           "Number of Purchases": [Total_items], 
                           "Total Revenue": [Total_revenue]})

summary_df
```

```python
#Percentage and Count of Players

gender_data = purchase_data_df.groupby("Gender")

gender_ratio = gender_data.nunique()["SN"]


percent_calculate = round((gender_ratio/Player_count), 4)*100

percent = percent_calculate.astype(str) + '%'


#gender_ratio

#percent


df = pd.DataFrame({"Percentage of Players": percent, "Total Count": gender_ratio})

df
```

```python
### Purchasing Analysis (Gender)

male_purchase_df = purchase_data_df[purchase_data_df['Gender'] == 'Male']

purchase_df_female = purchase_data_df[purchase_data_df['Gender'] == 'Female']

purchase_df_others = purchase_data_df[purchase_data_df['Gender'] == 'Other / Non-Disclosed']


purchase_count_male = len(male_purchase_df)
purchase_count_female = len(purchase_df_female)

purchase_count_others = len(purchase_df_others)

Purchase_price_male = male_purchase_df["Price"].sum()
Purchase_price_female = purchase_df_female["Price"].sum()
Purchase_price_others = purchase_df_others["Price"].sum()


Average_purchase_price_male = round(Purchase_price_male/purchase_count_male, 2)
Average_purchase_price_female = round(Purchase_price_female/purchase_count_female, 2)
Average_purchase_price_others = round(Purchase_price_others/purchase_count_others, 2)


number_purchase_df = gender_data["Purchase ID"].count()
average_purchase_df = round((gender_data["Price"].mean()), 2)
total_purchase_df = gender_data["Price"].sum()
per_person_average = round((total_purchase_df/gender_ratio), 2) 
#per_person_average

#print("Purchase count of games for males is", purchase_count_male)
#print("Total Purchase value of games for males is", Purchase_price_male)
#print("Average Purchase Price of games for males is", Average_purchase_price_male)

#print("Purchase count of games for females is", purchase_count_female )
#print("Total Purchase value of games for females is", Purchase_price_female)
#print("Average Purchase Price of games for females is", Average_purchase_price_female)


#print("Purchase count of games for others/Non-Disclosed is", purchase_count_others )
#print("Total Purchase value of games for others/Non-Disclosed is", Purchase_price_others)
#print("Average Purchase Price of games for others/Non-Disclosed is", Average_purchase_price_others)

Purchasing_Analysis_df = pd.DataFrame({"Purchase Count": number_purchase_df,
                                      "Average Purchase Price": average_purchase_df, 
                                      "Total Purchase Value": total_purchase_df,
                                      "Avg Total Purchase per Person": per_person_average})
                                   
Purchasing_Analysis_df                                     
```

```python
### Age Demographics

#Creating bins for the data

bin_df = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 100]
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#cutting based on the bins created and introducing new lables
purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"],bin_df, labels=bin_names)

#purchase_data_df

#grouping by age groups

age_group_df = purchase_data_df.groupby("Age Group")
age_group_df.head()

age_count = age_group_df["SN"].nunique()
age_count


percentage_df = round(((age_count/Player_count) * 100), 2)
percentage_df

age_demographics = pd.DataFrame({"Percentage of Players": percentage_df, "Total Count": age_count})
age_demographics
```

```python
### By age analysis


#calculating the purchase count
purchase_count_byage = age_group_df["Purchase ID"].count()

#calculating total value
total_value_df = age_group_df["Price"].sum()

#calculating avg purchase price by age
avg_price_byage = round(age_group_df["Price"].mean(), 2)

#print(purchase_count_byage)
#print(total_value_df)
#print(avg_price_byage)


#finding avg price pp = total avg/total age group numbers
avg_purchase_per_person_byage = round((total_value_df/age_count), 2)

#print(avg_purchase_per_person_byage)

byage_df = pd.DataFrame({"Purchase Count": purchase_count_byage,
                                 "Average Purchase Price": avg_price_byage,
                                 "Total Purchase Value":total_value_df,
                                 "Average Purchase Total per Person": avg_purchase_per_person_byage})

byage_df
```

```python
#sorting by top spender

#grouping by "SN"
sorting_by_spender = purchase_data_df.groupby("SN")

#sorting_by_spender.head()

#calculating the purchase count
count_byspender = sorting_by_spender["Purchase ID"].count()

#calculating avg purchase price by name 
avg_price_byspender = round(sorting_by_spender["Price"].mean(), 2)

#calculating total value
total_price_byspender = sorting_by_spender["Price"].sum()

spenders_df = pd.DataFrame({"Purchase Count": count_byspender,
                             "Average Purchase Price": avg_price_byspender,
                             "Total Purchase Value":total_price_byspender})
spenders_df

sorted_byspender = spenders_df.sort_values(["Total Purchase Value"], ascending = False)
sorted_byspender.head()
```

```python
### Most Popular Items

######grouping by "item" related info and then sorting
items_df = purchase_data_df[["Item ID", "Item Name", "Price"]]
sorted_items = items_df.groupby(["Item ID","Item Name"])


#####calculating total value
total_price_popular = (sorted_items["Price"].sum())

#####calculating the purchase count
count_popular = sorted_items["Price"].count()

#####calculating price per item
price_per_item_popular = total_price_popular/count_popular

most_popular_items = pd.DataFrame({"Purchase Count": count_popular, 
                                   "Item Price": price_per_item_popular,
                                   "Total Purchase Value":total_price_popular})

#most_popular_items - going by the assumption that number of "purchase count" would define the popularity of an item

sorted_most_popular_items = most_popular_items.sort_values(["Purchase Count"], ascending=False)
sorted_most_popular_items.head()
```

```python
### Most Profitable Items - Most profitable item can be sorted based on the "total purchase value" calculated in the last dataframe

most_profitable_df = sorted_most_popular_items.sort_values(["Total Purchase Value"],ascending=False)
most_profitable_df.head()
```
