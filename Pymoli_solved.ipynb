```python
### Age Demographics

#Creating bins for the data

bin_df = [0, 9.99, 14.99, 19.99, 24.99, 29.99, 34.99, 39.99, 100]
bin_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#cutting based on the bins created and introducing new lables
purchase_data_df["Age Group"] = pd.cut(purchase_data_df["Age"],bin_df, labels=bin_names)

#purchase_data_df

#grouping by age groups

age_group_df = purchase_data_df.groupby("Age Group")
age_group_df.head()

age_count = age_group_df["SN"].nunique()
age_count


percentage_df = round(((age_count/Player_count) * 100), 2)
percentage_df

age_demographics = pd.DataFrame({"Percentage of Players": percentage_df, "Total Count": age_count})
age_demographics
```

```python
### By age analysis


#calculating the purchase count
purchase_count_byage = age_group_df["Purchase ID"].count()

#calculating total value
total_value_df = age_group_df["Price"].sum()

#calculating avg purchase price by age
avg_price_byage = round(age_group_df["Price"].mean(), 2)

#print(purchase_count_byage)
#print(total_value_df)
#print(avg_price_byage)


#finding avg price pp = total avg/total age group numbers
avg_purchase_per_person_byage = round((total_value_df/age_count), 2)

#print(avg_purchase_per_person_byage)

byage_df = pd.DataFrame({"Purchase Count": purchase_count_byage,
                                 "Average Purchase Price": avg_price_byage,
                                 "Total Purchase Value":total_value_df,
                                 "Average Purchase Total per Person": avg_purchase_per_person_byage})

byage_df
```

```python
#sorting by top spender

#grouping by "SN"
sorting_by_spender = purchase_data_df.groupby("SN")

#sorting_by_spender.head()

#calculating the purchase count
count_byspender = sorting_by_spender["Purchase ID"].count()

#calculating avg purchase price by name 
avg_price_byspender = round(sorting_by_spender["Price"].mean(), 2)

#calculating total value
total_price_byspender = sorting_by_spender["Price"].sum()

spenders_df = pd.DataFrame({"Purchase Count": count_byspender,
                             "Average Purchase Price": avg_price_byspender,
                             "Total Purchase Value":total_price_byspender})
spenders_df

sorted_byspender = spenders_df.sort_values(["Total Purchase Value"], ascending = False)
sorted_byspender.head()
```

```python
### Most Popular Items

######grouping by "item" related info and then sorting
items_df = purchase_data_df[["Item ID", "Item Name", "Price"]]
sorted_items = items_df.groupby(["Item ID","Item Name"])


#####calculating total value
total_price_popular = (sorted_items["Price"].sum())

#####calculating the purchase count
count_popular = sorted_items["Price"].count()

#####calculating price per item
price_per_item_popular = total_price_popular/count_popular

most_popular_items = pd.DataFrame({"Purchase Count": count_popular, 
                                   "Item Price": price_per_item_popular,
                                   "Total Purchase Value":total_price_popular})

#most_popular_items - going by the assumption that number of "purchase count" would define the popularity of an item

sorted_most_popular_items = most_popular_items.sort_values(["Purchase Count"], ascending=False)
sorted_most_popular_items.head()
```

```python
### Most Profitable Items - Most profitable item can be sorted based on the "total purchase value" calculated in the last dataframe

most_profitable_df = sorted_most_popular_items.sort_values(["Total Purchase Value"],ascending=False)
most_profitable_df.head()
```
